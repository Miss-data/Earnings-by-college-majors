### 6) Hypothesis test:
#### One Way Anova test.
Here below i am going to run Anova one way test and check if there is difference in major categories average salaries and if it is statisticaly significant. 

- Status Quo: The average earnings of the major categories are equals
- Alternatively: The average earnings of the major categories are differents

```python
model = ols('Median~ C(Major_category)', data=EM).fit()          
aov_table = sm.stats.anova_lm(model, typ=2)
```

```python
def anova_table(aov):
    aov['mean_sq'] = aov[:]['sum_sq']/aov[:]['df']

    aov['eta_sq'] = aov[:-1]['sum_sq']/sum(aov['sum_sq'])

    aov['omega_sq'] = (aov[:-1]['sum_sq']-(aov[:-1]['df']*aov['mean_sq'][-1]))/(sum(aov['sum_sq'])+aov['mean_sq'][-1])

    cols = ['sum_sq', 'df', 'mean_sq', 'F', 'PR(>F)', 'eta_sq', 'omega_sq']
    aov = aov[cols]
    return aov

anova_table(aov_table)
```

From the above test result, the F-statistic is 13.99 greater than zero, so there is a difference between the major categories average earnings, The P-value is 4.787e-22 less than alpha 0.05 so we reject the Status Quo and conclude that there is a significant difference between major categories average earnings. The eta_sq is 57.21 % so the model explain 57.21 % of the variability in the response data (Average earnings) around it means. 

#### Tukey Honestly Significant Difference(HSD)
I am going to run HSD test to see which major category average earning significantly differ from each other.

```python
comp = mc.MultiComparison(EM['Median'], EM['Major_category'])
post_hoc_res = comp.tukeyhsd()
post_hoc_res.summary()
```

From the above results, Engineering is the major category that has an average earning that significanlty differ from most of the other major categories. In most of the case the confidence interval between the lower and upper does not contain zero and the null hypothesis is rejected.
